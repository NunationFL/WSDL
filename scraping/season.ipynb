```python
import requests
from bs4 import BeautifulSoup
import time


# print(page.content)
```

```python
headers = {'User-Agent': 'Mozilla/5.0'}
page = requests.get("https://www.zerozero.pt/edicao.php?id_edicao=165864", timeout=5, headers=headers)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup)
```

```python
b = soup.find_all("tbody")
# print(b[1])

clubs = {}
player_info = {}

for x in b[1]:
    # print(x.prettify())
    name = x.find("a")
    values = x.find_all("td")
    clubs[name.text] = [val.text for val in values]
    clubs[name.text] += [name['href']]
    time.sleep(1)
    club_page = requests.get("https://www.zerozero.pt"+name['href'], timeout=5, headers=headers)
    soup2 = BeautifulSoup(club_page.content, 'html.parser')
    team_html = soup.find_all(id="team_squad")

    team_players = {}
    for team in team_html[0]:
        players_per_position = team.find_all("div", class_="staff")
        for player in players_per_position:
            number = player.find("div", class_="number").text
            player_name = player.find("div", class_="name").find("div", class_="text").text
            age = player.find("div", class_="name").find("span").text
            link = player.find("div", class_="name").find("div", class_="text").find("a")["href"]
            team_players[number] = [player_name, age, link]
    clubs[name.text] += [team_players]
    print(clubs)


```
